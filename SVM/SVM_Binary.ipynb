```python
import os
import numpy as np
import pandas as pd
from  sklearn.svm import SVC
```

```python
data = pd.read_csv('/content/cardio_train.csv',delimiter=';')
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
```

```python
y = data['cardio']
data.drop(['cardio'],axis =1, inplace=True)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(data,y ,test_size = 0.3, random_state=12)
```

```python
X_train
```

```python
svm_clf = make_pipeline(StandardScaler(), SVC(decision_function_shape='ovo'))
svm_clf.fit(X_train, Y_train)
```

```python
y_pred = svm_clf.predict(X_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test, y_pred))
```
