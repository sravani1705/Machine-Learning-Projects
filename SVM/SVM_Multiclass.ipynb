```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('/content/healthcare_dataset.csv',delimiter=',')
data.head()
```

```python
data.info()
```

```python
data.drop(['Name','Doctor','Hospital','Insurance Provider','Room Number','Date of Admission','Discharge Date'],axis =1, inplace=True)
data.head()
```

```python
for cols in data.columns:
  print(cols)
  print(data[cols].unique())
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
```

```python
le = LabelEncoder()
col_data = ['Gender','Blood Type','Medical Condition','Billing Amount','Admission Type','Medication','Test Results']
for col in col_data:
  data[col] = le.fit_transform(data[col])
data.head()
```

```python
y = data['Test Results']
data.drop(['Test Results'],axis=1, inplace =True)
y.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(data,y,test_size=0.3,random_state = 17)
X_train
```

```python
dec_clf = DecisionTreeClassifier(max_depth=10)
dec_clf.fit(X_train,Y_train)
```

```python
y_pred = dec_clf.predict(X_test)
accuracy_score(Y_test,y_pred)
```

```python
from sklearn.pipeline import make_pipeline
svm_clf = make_pipeline(StandardScaler(),SVC(decision_function_shape='ovr',gamma = 'auto'))
svm_clf.fit(X_train, Y_train)
```

```python
y_pred1 = svm_clf.predict(X_test)
accuracy_score(Y_test,y_pred1)
```
